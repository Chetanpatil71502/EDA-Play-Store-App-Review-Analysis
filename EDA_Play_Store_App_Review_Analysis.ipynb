## 3. ***Data Wrangling***

### Cleaning Installs Column

````python
# removing '+' and ',' from install and coverting to integers
playstore_df['Installs'] = playstore_df['Installs'].str.replace('[+,]', '', regex=True)

playstore_df['Installs'] = pd.to_numeric(playstore_df['Installs'], errors='coerce')

playstore_df['Installs'] = playstore_df['Installs'].fillna(0).astype(int)
````

### Cleaning Price Column

````python
# Removing '$' symbol and coverting to float
playstore_df['Price'] = playstore_df['Price'].str.replace('$', '', regex=False)
playstore_df['Price'] = pd.to_numeric(playstore_df['Price'], errors='coerce')
````

### Cleaning size Column

````python
# Converting Size into numeric (handling MB and KB)

def convert_size(size):
    if size == 'Varies with device':
        return np.nan
    elif 'M' in size:
        return float(size.replace('M', ''))
    elif 'k' in size:
        return float(size.replace('k', '')) / 1024
    else:
        return np.nan

playstore_df['Size'] = playstore_df['Size'].apply(convert_size)
````

### Handling Outliers in Rating

````python
# Removing rows where rating is less than 5
playstore_df=playstore_df[playstore_df['Rating']<=5]
````

### Converting Reviews Column

````python
# converting Reviews to numeric
playstore_df['Reviews']=pd.to_numeric(playstore_df['Reviews'],errors='coerce')
````
