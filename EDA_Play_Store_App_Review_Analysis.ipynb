#### Handling Missing Values

````python
# For Play Store dataset

# Filling missing ratings with median (better than mean for skewed data)
playstore_df['Rating']=playstore_df['Rating'].fillna(playstore_df['Rating'].median())

# Filling missing type with most frequent value
playstore_df['Type']=playstore_df['Type'].fillna(playstore_df['Type'].mode()[0])
````

````python
# For Review Dataset

# Droping rows where review text is missing
reviews_df.dropna(inplace=True)
````
